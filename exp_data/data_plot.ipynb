```julia
dataframe = CSV.read("hiteijo_accuracy_1K_100K_for_graph.csv", delim=',');
```

```julia
deleterows!(dataframe, [i for i=1:2:size(dataframe)[1]])
symbols = names(dataframe)
symbols = symbols[2:length(symbols)]
labels = [string(s) for s in symbols]
labels[3] = "RS opt"
labels[2] = "Thompson Sampling"
plot(dataframe[1], convert(Array, dataframe[symbols]),xlabel="step",ylabel="regret", label=labels, legend=:right, fmt = :png,
style=:auto,w=3, legendfont=font(11))
```

```julia
using Statistics
```

```julia
non_st = CSV.read("hiteijo_accuracy_1K_100K.csv", delim=',', datarow=99901, )
st = CSV.read("teijo_accuracy_1K_100K.csv", delim=',', datarow=99901)
deleterows!(non_st, [101])
deleterows!(st, [101]);
```

```julia
non_st_labels = names(non_st)[2:end]
st_labels = names(st)[2:end]

means = zeros(2, length(st_labels))
for i=1:length(st_labels)
    means[1, i] = mean(st[st_labels[i]]) 
    means[2, i] = mean(non_st[non_st_labels[i]]) 
end
```

```julia
plot(means ,　label=st_labels, st=:scatter)
#means
```
